```python
import pandas as pd
ser1 = pd.Series(['Rukshan', 'Prasadi', 'Gihan', 'Hansana'], index=['Ru', 'Pr', 'Gi', 'Ha'])
ser2 = pd.Series([25, 25, 26, 24], index=['Ru', 'Pr', 'Gi', 'Ha'])
ser3 = pd.Series([85, 90, 70, 80], index=['Ru', 'Pr', 'Gi', 'Ha'])
a_dict = {'name': ser1, 'age': ser2, 'marks': ser3}
df = pd.DataFrame(a_dict)
print("Initial DataFrame:\n", df, "\n")

highest = df[df['marks'] == df['marks'].max()]
print("Student with highest marks:\n", highest, "\n")

df.loc['Th'] = ['Tharushi', 23, 88]
print("After adding Tharushi:\n", df, "\n")

age_25_plus = df[df['age'] >= 25]
print("Students aged 25 or older:\n", age_25_plus, "\n")

df.loc[df['marks'] < 80, 'marks'] += 10
print("After adding 10 bonus marks:\n", df, "\n")

avg_marks = df['marks'].mean()
print("Average Marks after bonus:", avg_marks)
```

```python
import pandas as pd
path = '/content/drive/My Drive/OnlineRetail.csv'
df = pd.read_csv(path, encoding='ISO-8859-1')
print("First 5 rows:")
print(df.head())
print("\nLast 5 rows:")
print(df.tail())
print("\nDataFrame Info:")
print(df.info())
print("\nDataFrame Shape:")
print(df.shape)
print("\nColumn Data Types:")
print(df.dtypes)
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
df = df.dropna(subset=['CustomerID'])
df['Description'] = df['Description'].fillna('Unknown')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce', infer_datetime_format=True)
df = df.dropna(subset=['InvoiceDate'])
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
print("\nNumber of duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()
df['Revenue'] = df['Quantity'] * df['UnitPrice']
df_outliers = df[df['Quantity'] > 1000]
df = df.sort_values(by='Revenue', ascending=False)
product_info = pd.DataFrame({
    'StockCode': ['10000', '10001', '10002'],
    'Category': ['Furniture', 'Electronics', 'Decor'],
    'Supplier': ['SupplierA', 'SupplierB', 'SupplierC']
})
df = pd.merge(df, product_info, on='StockCode', how='left')
country_group = df.groupby('Country').agg(
    total_revenue=('Revenue', 'sum'),
    mean_revenue=('Revenue', 'mean'),
    order_count=('InvoiceNo', 'nunique')
).reset_index()
price_bins = [0, 5, 20, df['UnitPrice'].max()]
price_labels = ['cheap', 'medium', 'expensive']
df['Price_Range'] = pd.cut(df['UnitPrice'], bins=price_bins, labels=price_labels)

price_group = df.groupby('Price_Range').agg(
    purchase_count=('InvoiceNo', 'count'),
    avg_revenue=('Revenue', 'mean')
).reset_index()

corr_matrix = df[['Quantity','UnitPrice','Revenue']].corr()
print("\nCorrelation matrix:\n", corr_matrix)

df = df.rename(columns={'StockCode':'Product_ID', 'UnitPrice':'Price'})
df['Description_lower'] = df['Description'].str.lower()
df_glass = df[df['Description_lower'].str.contains('glass')]

print("\nFiltered products containing 'glass':")
print(df_glass.head())

print("\nDescriptive Statistics:")
print(df[['Quantity', 'Price', 'Revenue']].describe())

print("\nValue counts by Country:")
print(df['Country'].value_counts())

if 'Category' in df.columns:
    print("\nValue counts by Category:")
    print(df['Category'].value_counts())

high_revenue_germany = df.query("Revenue > 1000 and Country == 'Germany'")
quantity_between_10_100 = df[df['Quantity'].between(10, 100)]

print("\nHigh Revenue Orders in Germany:")
print(high_revenue_germany.head())

print("\nOrders with Quantity between 10 and 100:")
print(quantity_between_10_100.head())

export_path = '/content/drive/My Drive/OnlineRetail_Cleaned.csv'
df.to_csv(export_path, index=False)
print(f"\nCleaned dataset saved to: {export_path}")

```
