```python
import numpy as np
farmA = np.array([[12,13.5,11.8],[10.2,11,10.8],[14.1,14.5,13.9],[9.5,10.1,9.8]])
farmB = np.array([[8.5,9,8.9],[11.2,11.9,12.1],[13,12.7,13.4],[7.8,8,7.9]])
farmC = np.array([[15,15.5,15.2],[9.9,10.3,10],[12.5,12.8,12.6],[11,10.7,11.2]])

production_all = np.stack([farmA,farmB,farmC])
print(production_all)
print(production_all.shape, production_all.size, production_all.dtype, production_all.ndim)

# Aggregations
total_per_farm = production_all.sum((1,2))
mean_per_farm = production_all.mean((1,2))
std_per_farm  = production_all.std((1,2))
print("total_per_farm, mean_per_farm, std_per_farm",total_per_farm, mean_per_farm, std_per_farm)

# Weighted
hour_weights = np.array([0.5,1,0.75])
weighted_per_panel = production_all.dot(hour_weights)
print("weighted_per_panel",weighted_per_panel)

# Calibration
production_all[1] += 0.5

# Normalize
pf_mean = production_all.mean((1,2))
pf_std  = production_all.std((1,2))
norm = (production_all - pf_mean)/pf_std

# Ufuncs
expv, logv, sinv = np.exp(norm), np.log(np.abs(norm)+1), np.sin(norm)

# Logical masks
high_output = production_all > 20
print(np.any(high_output))

# Bitwise
alerts = np.tile(np.array([[1,0,1],[0,0,1],[1,1,0],[0,1,0]]),(3,1,1))
bit_inv = np.bitwise_not(alerts)

# Panel totals & maintenance
panel_totals = production_all.sum(2)
needs_maint = np.any(panel_totals<40,1)
print(panel_totals, needs_maint)

# KPIs
overall_total = production_all.sum()
peak_idx = np.unravel_index(np.argmax(production_all), production_all.shape)
avg_panel = panel_totals.mean(0)
print(overall_total, production_all[peak_idx], peak_idx, avg_panel)
```
