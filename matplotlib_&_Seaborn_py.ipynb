```python


print("Shape (rows, columns):", df.shape)
print("\nColumn Info:\n")
print(df.info())
print("\nSummary Statistics:\n")
print(df.describe())
print("\nMissing Values:\n", df.isnull().sum())



# Count duplicates before removing
dup_count = df.duplicated().sum()
print("Duplicate rows found:", dup_count)
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
print("Rows removed:", dup_count)



df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
invalid_dates = df['Date'].isna().sum()
print("Invalid/Missing dates:", invalid_dates)
df = df.sort_values(by='Date')



numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
print("Missing values in numeric columns:\n", df[numeric_cols].isnull().sum())
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print("\nMissing values after imputation:\n", df[numeric_cols].isnull().sum())



```

```python

from scipy.stats import pearsonr

#Duration vs Calories
plt.figure(figsize=(7,5))
plt.scatter(df['Duration'], df['Calories'], color='blue', alpha=0.6, label='Data points')

# Linear trend line
m, b = np.polyfit(df['Duration'], df['Calories'], 1)
plt.plot(df['Duration'], m*df['Duration'] + b, color='red', linestyle='--', label='Trend line')

#Pearson correlation
corr, _ = pearsonr(df['Duration'].dropna(), df['Calories'].dropna())
plt.title(f'Duration vs Calories (r = {corr:.2f})')
plt.xlabel('Duration (minutes)')
plt.ylabel('Calories')
plt.legend()
plt.grid(True)
plt.show()


#pie chart
bins = [0, 120, 140, 160, df['Maxpulse'].max()]
labels = ['<=120', '121-140', '141-160', '>160']
df['MaxpulseRange'] = pd.cut(df['Maxpulse'], bins=bins, labels=labels, include_lowest=True)
counts = df['MaxpulseRange'].value_counts().sort_index()

plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['lightblue','lightgreen','orange','red'], explode=[0.05]*4)
plt.title("Maxpulse Distribution")
plt.savefig("/content/drive/MyDrive/pie_maxpulse_ranges.png", bbox_inches='tight')
plt.show()


#hist
plt.figure(figsize=(7,5))
plt.hist(df['Calories'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Calories")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("/content/drive/MyDrive/hist_calories.png", bbox_inches='tight')
plt.show()

# Skewness comment
skewness = df['Calories'].skew()
print(f"Calories histogram skewness: {skewness:.2f}")
if skewness > 0:
    print("Distribution is positively skewed (long tail on the right).")
elif skewness < 0:
    print("Distribution is negatively skewed (long tail on the left).")
else:
    print("Distribution is roughly symmetric.")


#multi lone scatter

plt.figure(figsize=(14,5))
# Left: Pulse over Date
plt.subplot(1,2,1)
plt.plot(df['Date'], df['Pulse'], 'o-', color='purple', label='Pulse')
plt.title("Pulse Over Time")
plt.xlabel("Date")
plt.ylabel("Pulse")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Right: Duration vs Maxpulse
plt.subplot(1,2,2)
plt.scatter(df['Duration'], df['Maxpulse'], color='teal', alpha=0.6)
plt.title("Duration vs Maxpulse")
plt.xlabel("Duration (minutes)")
plt.ylabel("Maxpulse")
plt.grid(True)

plt.tight_layout()
plt.savefig("/content/drive/MyDrive/subplots_pulse_duration_maxpulse.png", bbox_inches='tight')
plt.show()
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'])
plt.title('KDE Plot - Smooth Distribution of Total Bill')
plt.show()


```

```python
# Displot (Histogram + KDE)
sns.displot(tips['total_bill'], kde=True)
plt.title('Displot - Total Bill Distribution')
plt.show()
```

```python
# Bar Plot - Average Tip by Gender
sns.barplot(x='sex', y='tip', data=tips)
plt.title('Bar Plot - Average Tip by Gender')
plt.show()
```

```python
# Count Plot - Frequency of Days
sns.countplot(x='day', data=tips)
plt.title('Count Plot - Frequency of Days')
plt.show()
```

```python
# Box Plot - Total Bill by Day
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot - Total Bill by Day')
plt.show()
```

```python
# Violin Plot - Distribution + Summary
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot - Total Bill by Day')
plt.show()
```

```python
# Scatter Plot - Total Bill vs Tip
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.title('Scatter Plot - Total Bill vs Tip')
plt.show()
```

```python
# Line Plot - Tip vs Party Size (useful for trend-like visuals)
sns.lineplot(x='size', y='tip', data=tips)
plt.title('Line Plot - Tip vs Party Size')
plt.show()
```

```python
# Relplot with Facets
sns.relplot(x='total_bill', y='tip', hue='sex', col='day', data=tips)
plt.suptitle('Relplot - Tip vs Total Bill by Day & Gender', y=1.02)
plt.show()
```

```python
# Select only numeric columns
iris_num = iris.select_dtypes(include=['float', 'int'])

# Compute correlation
corr = iris_num.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True,cmap='coolwarm')
plt.title("Heatmap - Correlation of Iris Features")
plt.show()
```

```python
# Select only numeric columns (exclude 'species')
iris_num = iris.select_dtypes(include=['float', 'int'])

# Compute correlation matrix
corr = iris_num.corr()

# Create the clustermap
sns.clustermap(corr, cmap='coolwarm', annot=True)
plt.suptitle('Clustermap - Grouped Correlations (Numeric Only)', y=1.02)
plt.show()
```
